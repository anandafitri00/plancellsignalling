```
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```
# load data
url = 'https://raw.githubusercontent.com/anandafitri00/plancellsignalling/main/plantCellSignaling.csv'
data = pd.read_csv(url)
# data = data.drop(columns=['*'])
data
```

```
data = data.drop(data[data.ABA=="*"].index)
```

```
# data = data.iloc[:1000]
data
```

```
# data.isnull().sum()
```

```
atr = data.columns
for j in range(len(atr)):
  data[atr[j]] = pd.to_numeric(data[atr[j]])

data.info()
```

```
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

data = data.applymap(encode_units)
```

```
data = data.sample(n=1000)
```

```
data.corr()
```

```
del data['ABA']
del data['AGB']
del data['KAP']
del data['KOUT']
del data['HATPase']
del data['DEPOLAR']
del data['ABI']
del data['AnionEM']
del data['ROS']
del data['GPA']
del data['ATRBOH']
del data['MALATE']
del data['cGMP']
del data['cADPR']
del data['IP3']
del data['PA']
del data['S1P']
del data['NOS']
del data['GC']
del data['KEV']
del data['RAC']
del data['RCN']
del data['NIA12']
del data['PLC']
del data['InsPK']
del data['IP6']
del data['ADPRc']
```

```
frequent_itemsets = apriori(data, min_support=0.07, use_colnames=True)
```

```
frequent_itemsets.head()
```

```
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```
rules.max()
```
